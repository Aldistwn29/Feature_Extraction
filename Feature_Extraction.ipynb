```python
import pandas as pd
```

```python
dataset = pd.read_csv("/content/sample_data/california_housing_train.csv")
```

```python
dataset.head()
```

# Math opration

```python
other_rooms = dataset["total_rooms"] - dataset["total_bedrooms"]
other_rooms
```

```python
# menggunakan Method
other_rooms = dataset["population"].div(dataset["households"])
other_rooms
```

```python
kepadatan = dataset["population"].div(dataset["households"])
kepadatan
```

```python
# Memasukan other room dalam dataset
dataset["other_rooms"] = other_rooms
dataset
```

```python
# Menggunakan .assign
dataset = dataset.assign(kepadatan = kepadatan)
dataset
```

```python
# Menggunakan insert
dataset.insert(2, "price_income_ratio", dataset["median_house_value"] / dataset["median_income"])
dataset
```

```python
kepadatan_ruang = dataset["total_rooms"] / dataset["households"]
kepadatan_ruang
```

```python
kepadatan_ruang.pow(2)
```

```python
# menggunakan metode insert
dataset.insert(0, "kepadatan_ruang_kuadrat", kepadatan_ruang.pow(2))
dataset
# menggunkan metode insert bisa di atur urutannya secara bebas dengan
# menyantunkan indek
```
