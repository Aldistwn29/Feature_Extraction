```python
import pandas as pd
```

```python
dataset = pd.read_csv("/content/sample_data/california_housing_train.csv")
```

```python
dataset.head()
```

# Math opration

```python
other_rooms = dataset["total_rooms"] - dataset["total_bedrooms"]
other_rooms
```

```python
# menggunakan Method
other_rooms = dataset["population"].div(dataset["households"])
other_rooms
```

```python
# slicing
kepadatan = dataset["population"].div(dataset["households"])
kepadatan
```

```python
# slicing
dataset["other_rooms"] = other_rooms
dataset
```

```python
# Menggunakan .assign
dataset = dataset.assign(kepadatan = kepadatan)
dataset
```

```python
# Menggunakan insert
dataset.insert(2, "price_income_ratio", dataset["median_house_value"] / dataset["median_income"])
dataset
```

```python
kepadatan_ruang = dataset["total_rooms"] / dataset["households"]
kepadatan_ruang
```

```python
kepadatan_ruang.pow(2)
```

```python
# menggunakan metode insert
dataset.insert(0, "kepadatan_ruang_kuadrat", kepadatan_ruang.pow(2))
dataset
# menggunkan metode insert bisa di atur urutannya secara bebas dengan
# menyantunkan indek
```

```python
import seaborn as sns
```

```python
mpg_dataset = sns.load_dataset("mpg")
```

```python
sns.get_dataset_names()
```

```python
mpg_dataset["origin"].unique()
```

```python
# Menumpuk velue di kolom yg di jadikan prameter
mpg_dataset.loc[mpg_dataset["origin"] == "usa", "origin"] = "america"
mpg_dataset
```

```python
# Membuat kolom baru
mpg_dataset.loc[mpg_dataset["origin"] == "usa", "new_origin"] = "america"
mpg_dataset
```

```python
mpg_dataset["new_origin"] = mpg_dataset["origin"].replace("japan", "asia")
mpg_dataset.query("origin == 'asia'")
```

```python
mpg_dataset["new_origin"] = mpg_dataset["origin"].map({"america":"usa",
                                                       "asia":"japan",
                                                       "europe":"europe"
})
mpg_dataset
```

```python
mpg_dataset = mpg_dataset.rename(columns ={"origin":"country"})
mpg_dataset
```

```python
# menghapus kolom
mpg_dataset = mpg_dataset.drop(columns = "name")
mpg_dataset
```

```python
# tanpa mendefinisikan variabel
mpg_dataset.drop(columns = "new_origin", inplace = True)
```
